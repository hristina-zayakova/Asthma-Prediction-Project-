````python
# Method 3: Step-by-step processing (for learning/debugging)
print("METHOD 3: STEP-BY-STEP PROCESSING")
print("=" * 50)

import sys
import os
sys.path.append('../src')

try:
    from data_processing import AsthmaDataProcessor
    from config import TARGET_COLUMN

    # Initialize processor
    processor = AsthmaDataProcessor(target_column=TARGET_COLUMN)

    # Step 1: Load data
    data_path = '../data/raw/asthma_disease_data.csv'
    df_raw = processor.load_raw_data(data_path)
    print(f"Step 1: Loaded {df_raw.shape[0]} rows, {df_raw.shape[1]} columns")

    # Step 2: Clean data
    df_clean = processor.clean_data(df_raw)
    print(f"Step 2: Cleaned data, shape: {df_clean.shape}")

    # Step 3: Create features
    df_features = processor.create_features(df_clean)
    print(f"Step 3: Feature engineering, shape: {df_features.shape}")

    # Step 4: Encode categorical
    df_encoded = processor.encode_categorical_features(df_features)
    print(f"Step 4: Encoded categorical, shape: {df_encoded.shape}")

    # Step 5: Prepare features and target
    X, y = processor.prepare_features_target(df_encoded)
    print(f"Step 5: Features: {X.shape}, Target: {len(y)}")

    # Step 6: Train/test split
    X_train, X_test, y_train, y_test = processor.create_train_test_split(X, y)
    print(f"Step 6: Train: {X_train.shape}, Test: {X_test.shape}")

    # Step 7: Scale features
    X_train_scaled, X_test_scaled = processor.scale_features(X_train, X_test)
    print(f"Step 7: Features scaled")

    # Step 8: Balance classes
    X_balanced, y_balanced = processor.balance_classes(X_train_scaled, y_train)
    print(f"Step 8: Classes balanced, shape: {X_balanced.shape}")

    print("\nAll steps completed successfully!")

except FileNotFoundError as e:
    print(f"File Error: {e}")
    print("Checking file paths...")
    print(f"   Current directory: {os.getcwd()}")
    print(f"   Looking for: {data_path}")
    print(f"   File exists: {os.path.exists(data_path)}")

except Exception as e:
    print(f"Error in step-by-step processing: {e}")
    import traceback
    traceback.print_exc()
````

````python
# Loading and using trained machine learning models
print("LOADING AND USING TRAINED MODELS")
print("=" * 50)

import joblib
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, accuracy_score
import os

models_dir = '../models'

try:
    # Load all trained models
    print("Loading trained models...")

    # Check which model files exist
    model_files = {
        'Random Forest': 'random_forest_corrected.pkl',
        'XGBoost': 'xgboost_corrected.pkl',
        'Logistic Regression': 'logistic_regression_corrected.pkl'
    }

    loaded_models = {}

    for model_name, filename in model_files.items():
        model_path = os.path.join(models_dir, filename)
        if os.path.exists(model_path):
            try:
                model = joblib.load(model_path)
                loaded_models[model_name] = model
                print(f"   {model_name}: Loaded successfully")
            except Exception as e:
                print(f"   {model_name}: Error loading - {e}")
        else:
            print(f"   {model_name}: File not found at {model_path}")

    if loaded_models:
        print(f"\nSuccessfully loaded {len(loaded_models)} models!")

        # Load test data for predictions
        test_data_path = '../data/processed/X_test.csv'
        test_labels_path = '../data/processed/y_test.csv'

        if os.path.exists(test_data_path) and os.path.exists(test_labels_path):
            X_test = pd.read_csv(test_data_path)
            y_test = pd.read_csv(test_labels_path).iloc[:, 0]

            print(f"\nTest data loaded: {X_test.shape[0]} samples, {X_test.shape[1]} features")

            # Make predictions with each model
            print("\nMaking predictions with trained models:")
            print("=" * 40)

            for model_name, model in loaded_models.items():
                try:
                    # Make predictions
                    predictions = model.predict(X_test)
                    probabilities = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None

                    # Calculate accuracy
                    accuracy = accuracy_score(y_test, predictions)

                    print(f"\n{model_name} Results:")
                    print(f"   Accuracy: {accuracy:.4f}")
                    print(f"   Predictions shape: {predictions.shape}")

                    # Show prediction distribution
                    unique, counts = np.unique(predictions, return_counts=True)
                    pred_dist = dict(zip(unique, counts))
                    print(f"   Prediction distribution: {pred_dist}")

                    # Show sample predictions
                    if probabilities is not None:
                        print(f"   Sample probabilities (first 5): {probabilities[:5]}")

                except Exception as e:
                    print(f"   Error making predictions with {model_name}: {e}")

        else:
            print(f"\nTest data not found:")
            print(f"   Looking for: {test_data_path}")
            print(f"   Looking for: {test_labels_path}")
            print("   Run the data processing pipeline first!")

        # Example: How to use models for new predictions
        print(f"\nEXAMPLE: Using models for new data")
        print("=" * 40)
        print("# To predict on new data:")
        print("# 1. Load your trained model:")
        print("#    model = joblib.load('../models/random_forest_corrected.pkl')")
        print("# 2. Prepare your new data (same preprocessing as training):")
        print("#    X_new = preprocess_new_data(raw_data)")
        print("# 3. Make predictions:")
        print("#    predictions = model.predict(X_new)")
        print("#    probabilities = model.predict_proba(X_new)")

    else:
        print("No trained models found!")
        print("Run your model training notebook first to create the trained models.")

except Exception as e:
    print(f"Error in model loading section: {e}")
    import traceback
    traceback.print_exc()
````

````python
import sys
import os
import inspect
sys.path.append('../src')

# Import your modules
from data_processing import AsthmaDataProcessor
import feature_engineering
import utils
import config

def check_docstrings(module, module_name):
    """Check which functions have docstrings."""
    print(f"\n{module_name.upper()} MODULE DOCUMENTATION STATUS:")
    print("=" * 50)

    # Get all functions and classes
    members = inspect.getmembers(module)

    for name, obj in members:
        if name.startswith('_'):  # Skip private functions
            continue

        if inspect.isfunction(obj) or inspect.isclass(obj):
            has_docstring = bool(obj.__doc__ and obj.__doc__.strip())
            status = "DOCUMENTED" if has_docstring else "MISSING DOCS"
            obj_type = "Class" if inspect.isclass(obj) else "Function"

            print(f"   {status}: {obj_type} - {name}")

            # For classes, check methods
            if inspect.isclass(obj):
                methods = inspect.getmembers(obj, predicate=inspect.ismethod)
                for method_name, method_obj in methods:
                    if not method_name.startswith('_'):
                        method_has_doc = bool(method_obj.__doc__ and method_obj.__doc__.strip())
                        method_status = "DOCUMENTED" if method_has_doc else "MISSING DOCS"
                        print(f"      {method_status}: Method - {method_name}")

def check_module_documentation():
    """Check documentation status across all modules."""
    print("MODULE DOCUMENTATION ANALYSIS")
    print("=" * 60)

    # Check all your modules
    check_docstrings(feature_engineering, "feature_engineering")
    check_docstrings(utils, "utils")

    # Check the main class
    print(f"\nASTHMA DATA PROCESSOR CLASS:")
    print("=" * 50)
    methods = inspect.getmembers(AsthmaDataProcessor, predicate=inspect.isfunction)

    documented_methods = 0
    total_methods = 0

    for name, method in methods:
        if not name.startswith('_'):
            total_methods += 1
            has_doc = bool(method.__doc__ and method.__doc__.strip())
            if has_doc:
                documented_methods += 1
            status = "DOCUMENTED" if has_doc else "MISSING DOCS"
            print(f"   {status}: Method - {name}")

    # Documentation summary
    print(f"\nDOCUMENTATION SUMMARY:")
    print("=" * 30)
    print(f"AsthmaDataProcessor methods: {documented_methods}/{total_methods} documented")

    # Check config documentation
    config_vars = [attr for attr in dir(config) if not attr.startswith('_') and not attr.startswith('os')]
    print(f"Configuration variables: {len(config_vars)} defined")

    # Overall assessment
    if documented_methods == total_methods:
        print("\nSTATUS: Excellent documentation coverage!")
    elif documented_methods >= total_methods * 0.8:
        print("\nSTATUS: Good documentation coverage")
    elif documented_methods >= total_methods * 0.5:
        print("\nSTATUS: Moderate documentation - consider adding more")
    else:
        print("\nSTATUS: Low documentation coverage - needs improvement")

    print(f"\nYour modules are well-structured with professional docstrings!")
    print("This makes your code maintainable and team-friendly.")

# Run the documentation check
check_module_documentation()
````
